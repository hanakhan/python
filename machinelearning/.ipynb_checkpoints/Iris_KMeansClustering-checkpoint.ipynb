For this project I will attempt to use KMeans Clustering to cluster the plants in the iris dataset into three distinct groups.

Usually the Kmeans algorithm is used when the labels are not present in the data. In this case, the labels (species) are included with the data set we are using for this project. ## The Data

Here is a description of the data taken from <a href='https://www.kaggle.com/uciml/iris'> Kaggle.com</a>: <br>
The Iris dataset was used in Fisher's classic 1936 paper, <i>The Use of Multiple Measurements in Taxonomic Problems.</i>

It includes three iris species with 50 samples each as well as some properties about each flower. One flower species is linearly separable from the other two, but the other two are not linearly separable from each other.

The columns in this dataset are:

Id<br>
SepalLengthCm<br>
SepalWidthCm<br>
PetalLengthCm<br>
PetalWidthCm<br>
Species

## K Means Cluster Creation

Now it is time to create the Cluster labels!

** Import KMeans from SciKit Learn.**

```python
from sklearn.cluster import KMeans
```

** Create an instance of a K Means model with 2 clusters.**

```python
kmeans = KMeans(n_clusters=3)
```

**Fit the model to all the data except for the species label.**

```python
kmeans.fit(iris.drop('species',axis=1))
```

** What are the cluster center vectors?**

```python
kmeans.cluster_centers_
```

**These are the predicted labels**

```python
kmeans.labels_
```

**Represent species with numerical values instead of species name for model evaluation purposes**

```python
iris['species'].replace(['setosa','virginica','versicolor'],[0,1,2],inplace=True)
```

```python
iris.head()
```

## Evaluation

With this dataset, we have the actual labels we can compare our results with. Let's visualize the clusters created by the K Means model.

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(iris['petal_width'],iris['petal_length'],c=kmeans.labels_,cmap='viridis')
ax2.set_title("Original")
ax2.scatter(iris['petal_width'],iris['petal_length'],c=iris['species'],cmap='viridis')
```

The model did a pretty good job in clustering the plants into separate species! Notice that the colors do not match between the plots (if it does, it's by chance!), which is expected since we wanted the model to purely split the data into 3 distinct clusters.
